```python
import math
import cv2
import numpy as np
import time

cam = cv2.VideoCapture(0)
cv2.namedWindow("Camera", cv2.WINDOW_KEEPRATIO)

lower_g = (51, 90, 120)
upper_g = (70, 255, 255)

#lower_b = (9, 240, 180)
#upper_b = (100, 255, 134)

lower_b = (80, 200, 150)
upper_b = (120, 280, 255)

lower_y = (20, 160, 170)
upper_y = (30, 255, 255)

lower_r = (0, 150, 140)
upper_r = (15, 255, 255)


def get_circle(hsv, lower, upper):
    mask = cv2.inRange(hsv, lower, upper)
    mask = cv2.erode(mask, None, iterations=2)
    mask = cv2.dilate(mask, None, iterations=2)

    cnts = cv2.findContours(mask.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)[-2]

    if len(cnts) > 0:
        c = max(cnts, key=cv2.contourArea)
        (curr_x, curr_y), radius = cv2.minEnclosingCircle(c)
        if radius > 10:
            return (curr_x, curr_y), radius
    return None

while cam.isOpened():
    _, image = cam.read()
    blurred = cv2.GaussianBlur(image, (9, 9), 0)

    hsv = cv2.cvtColor(blurred, cv2.COLOR_BGR2HSV)

    circle_b = get_circle(hsv, lower_b, upper_b)
    circle_y = get_circle(hsv, lower_y, upper_y)
    circle_r = get_circle(hsv, lower_r, upper_r)
    circle_g = get_circle(hsv, lower_g, upper_g)

    if circle_b:
        cv2.circle(blurred, (int(circle_b[0][0]), int(circle_b[0][1])), int(circle_b[1]), (255, 255, 0), 3)
        cv2.circle(blurred, (int(circle_b[0][0]), int(circle_b[0][1])), 3, (255, 255, 0), 3)

    if circle_y:
        cv2.circle(blurred, (int(circle_y[0][0]), int(circle_y[0][1])), int(circle_y[1]), (0, 255, 255), 3)
        cv2.circle(blurred, (int(circle_y[0][0]), int(circle_y[0][1])), 3, (0, 255, 255), 3)

    if circle_r:
        cv2.circle(blurred, (int(circle_r[0][0]), int(circle_r[0][1])), int(circle_r[1]), (0, 0, 255), 3)
        cv2.circle(blurred, (int(circle_r[0][0]), int(circle_r[0][1])), 3, (0, 0, 255), 3)

    if circle_g:
        cv2.circle(blurred, (int(circle_g[0][0]), int(circle_g[0][1])), int(circle_g[1]), (0, 255, 0), 3)
        cv2.circle(blurred, (int(circle_g[0][0]), int(circle_g[0][1])), 3, (0, 255, 0), 3)

    cv2.imshow("Camera", blurred)

    key = cv2.waitKey(1)
    if key == ord('q'):
        break


cam.release()
cv2.destroyAllWindows()
```
